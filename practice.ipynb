```python
import pandas as pd
import numpy as np
from datetime import datetime
import calendar
```

```python
cust_details=pd.read_csv('Case_Study_Customer_Data.csv')
```

```python
order_data=pd.read_csv('Case_Study_Orders_Data.csv')
```

```python
order_data
```

```python
order_data['order_date']=pd.to_datetime(order_data['Order Date']).dt.date
```

```python
cust_details['Customer Created Date']=pd.to_datetime(cust_details['Customer Created Date']).dt.date
```

```python
cust_details['Customer Created Date']=cust_details['Customer Created Date'].astype(str)
```

```python
cust_details['created_month_number']=cust_details['Customer Created Date'].str.split("-").str[1]
```

```python
cust_details['created_month_name']=cust_details.month_number.replace(dic,regex=True)
```

```python
order_data['order_date']=order_data['order_date'].astype(str)
```

```python
order_data['order_month_number']=order_data['order_date'].str.split("-").str[1]
```

```python
dic={'01':'Jan','02':'Feb','03':'Mar','04':'Apr','05':'May','06':'Jun','07':'Jul','08':'Aug','09':'Sep','10':'Oct','11':'Nov','12':'Dec'}
```

```python
order_data['order_month_name']=order_data.month_number.replace(dic,regex=True)
```

```python
monthly_revenue=order_data.groupby(['order_month_name']).agg({'Order Amount':'sum'}).reset_index()
```

```python
monthly_revenue.columns
```

```python
import matplotlib.pyplot as plt
monthly_revenue.plot(x='order_month_name', y='Order Amount', kind='bar', legend=False)
plt.xlabel('Month')
plt.ylabel('Order Amount')
plt.title('Monthly Revenue')
```

```python
cust_count=order_data.groupby('order_month_name').size().reset_index()
cust_count.rename(columns={0:'count'},inplace=True)
```

```python
cust_count.columns
```

```python
cust_count.plot(x='order_month_name', y='count', kind='bar', legend=False)
plt.xlabel('Month')
plt.ylabel('Cust_count')
plt.title('Customer Counts')
```

```python
per_cust=monthly_revenue.merge(cust_count,on='order_month_name')
```

```python
per_cust['revenue_per_user']=per_cust['Order Amount']/per_cust['count']
```

```python
per_cust['revenue_per_user']=np.round(per_cust['revenue_per_user'], 
                       decimals = 2)
```

```python
per_cust=per_cust.reset_index()
```

```python
per_cust.plot(x='order_month_name', y='revenue_per_user', kind='bar', legend=False)
plt.xlabel('Month')
plt.ylabel('revenue')
plt.title('Revenue Per Customers')
```

```python
per_cust.set_index('order_month_name', inplace=True)
ax = per_cust.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Month')
plt.ylabel('Values')
plt.title('Monthly Performance Summary')
plt.legend(title='Metrics')
plt.show()
```

```python
ax = per_cust.plot.pie(subplots=True, figsize=(25, 15), autopct='%1.1f%%', startangle=90, legend=False)
plt.suptitle('Monthly Performance Summary')
plt.show()
```
